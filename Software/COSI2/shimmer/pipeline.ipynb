# **shimming pipeline**
## this script is used to shim OSI-type hallbach MRI magnets
### input: 
- stl model of a magnet
- map of B0 field measured with the magnetometer along a path inside the magnet
- configuration of shimming rings
- parameters of shimming magnets

### output:
- simulated shimmed field, homogeneity in ppm
- rotations of shim magnets in the rings

```python
# importing the classes
import b0
import plotter
import numpy as np


b0map = b0.b0()  # dummy b0 map generated
b0map.create_anomaly(x=0.6,y=-0.4,z=2.5,strength=15,length=0) # create a singular value at coordinates x,y,z

b0plotter = plotter.plotter(map_to_plot = b0map)
print(np.nanmax(b0map.magnetic_field))




```
